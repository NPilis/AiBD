W Pythonie bloki kodu (w tym również funkcje) są wyróżniane za pomocą wcięć. Do wizualizacji wyników w Pythonie używa się pakietu [Matplotlib](https://matplotlib.org/).

### Deklaracja pomocniczych funkcji i przedziałów

```python
f = lambda x : x**2 + 5

def pretty_plot(x, y, title, xlabel, ylabel, labels):
    plt.figure()
    plt.plot(x, y)
    plt.plot([x[0], x[-1]], [y[0], y[-1]], 'o', mfc='none', color='blue')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(labels=labels)
    plt.show()
    
X1 = np.linspace(-1, 1, 1000)
X2 = np.linspace(-6, 6, 1000)
X3 = np.linspace(0, 5, 1000)
```

```python
pretty_plot(X1, f(X1), "Wykres $f(x)=x^2+5$ dla\n$x>-1$ oraz $x<1$",
            "$x$", "$f(x)$", ["$f(x)$"])
```

```python
pretty_plot(X2, f(X2), "Wykres $f(x)=x^2+5$ dla\n$x>-6$ oraz $x<6$",
            "$x$", "$f(x)$", ["$f(x)$"])
```

```python
pretty_plot(X3, f(X3), "Wykres $f(x)=x^2+5$ dla\n$x>0$ oraz $x<5$",
            "$x$", "$f(x)$", ["$f(x)$"])
```
